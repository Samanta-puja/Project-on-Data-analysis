```python
#import python libraries, like- pandas,seaborn and matplotlib.
import pandas as pd  
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the Excel file into a DataFrame
file_path = 'Data Analysis - Caselet#2.xlsx'   # Replace with the actual path 
sheet_name = 'DifferentFormat'
df=pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
#head()- shows top 5 rows from data list, from given dataframe.
df.head()
```

```python
# Create a box plot for TPT
plt.figure(figsize=(13, 6))
plt.xticks(rotation=90)
sns.boxplot(x='Team Member', y='TPT', data=df)
plt.title('Box Plot of Total Processing Time (TPT) by Team Member')
plt.show()
```
